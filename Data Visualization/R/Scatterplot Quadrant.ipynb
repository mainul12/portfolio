````r
## 35 countries
````

````r
countries <- paste("Country ", 1:35)
````

````r
srh <- sample(x = 30:90,size = 35,replace = TRUE)
````

````r
le <- sample(70:90, size = 35, replace = TRUE)
````

````r
## create a dataframe
df <- data.frame(country_names = countries, 
                self_reported_health = srh, 
                life_expectancy = le)
````

````r
attach(df)
````

````r
X <- self_reported_health
Y <- life_expectancy
````

````r
X_description <- "Self-reported health"
Y_description <- "Life Expectancy"
````

````r
library(RColorBrewer)
````

````r
C1 <- brewer.pal(5, "PiYG")[5]
C2 <- brewer.pal(5, "PiYG")[4]
C3 <- brewer.pal(5, "PiYG")[1]
C4 <- brewer.pal(5, "PiYG")[2]
````

````r
Q1 <- subset(df, X > mean(X) & Y > mean(Y))
Q2 <- subset(df, X < mean(X) & Y > mean(Y))
Q3 <- subset(df, X < mean(X) & Y < mean(Y))
Q4 <- subset(df, X > mean(X) & Y < mean(Y))
````

````r
library(repr)
options(repr.plot.width = 12, repr.plot.height = 12)


par(mar = c(5, 5, 5, 5), omi = c(1, 1, 1, 1), las = 1)

plot(type = "n", xlab = X_description, 
     ylab = Y_description, X, Y, xlim = c(30, 90), ylim = c(70, 90), axes = F)
axis(1, col = par("bg"), col.ticks = "red", lwd.ticks = 0.5, tck=-0.025)
axis(2, col = par("bg"), col.ticks = "blue", lwd.ticks = 0.5, tck=-0.025)


symbols(Q1[, 2:3], bg = C1, circles = rep(1, nrow(Q1)), inches = 0.3, add=T, xpd = T, fg = "white")
symbols(Q2[, 2:3], bg = C2, circles = rep(1, nrow(Q2)), inches = 0.3, add=T, xpd = T, fg = "white")
symbols(Q3[, 2:3], bg = C3, circles = rep(1, nrow(Q3)), inches = 0.3, add=T, xpd = T, fg = "white")
symbols(Q4[, 2:3], bg = C4, circles = rep(1, nrow(Q4)), inches = 0.3, add=T, xpd = T, fg = "white")


text(Q2[, 2:3], as.matrix(Q2$country_names), cex = 1, pos = 1, offset = 0.7)
text(Q4[, 2:3], as.matrix(Q4$country_names), cex = 0.9, pos = 1, offset = 0.7)

abline(v=mean(X), col="black", lty=3)
abline(h=mean(Y), col="black", lty=3)
````
